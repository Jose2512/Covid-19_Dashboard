```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
import itertools
```

```python
file_to_load = "../../Resources/casos-asociados-a-covid-19.csv"

covid_cases_data = pd.read_csv(file_to_load)
covid_cases_data.head()
```

```python
clean_covid_data = covid_cases_data[covid_cases_data["positivo"]=="Positivo SARS-CoV-2"]
clean_covid_data = clean_covid_data[["ID_REGISTRO","SEXO","MUNICIPIO RESIDENCIA","FECHA INGRESO", "FECHA DEFUNCION","INTUBADO","EDAD","DIABETES","EPOC","ASMA","INMUNOSUPRESION","HIPERTENSION","CARDIOVASCULAR","OBESIDAD","RENAL CRONICA","TABAQUISMO"]]
display(clean_covid_data)
```

```python
fecha = clean_covid_data.groupby(["FECHA INGRESO"]).count().reset_index().rename(columns={'FECHA INGRESO':'FECHA'}).drop(['ID_REGISTRO','MUNICIPIO RESIDENCIA','SEXO','FECHA DEFUNCION','INTUBADO','EDAD','DIABETES','EPOC','ASMA','INMUNOSUPRESION','HIPERTENSION','CARDIOVASCULAR','OBESIDAD','RENAL CRONICA','TABAQUISMO'],axis=1)
delegaciones = clean_covid_data['MUNICIPIO RESIDENCIA'].unique().tolist()
delegaciones.pop(0)
fechas = clean_covid_data['FECHA INGRESO'].unique().tolist()
a = [fechas,delegaciones]
b=list(itertools.product(*a))
final_delelegacion_fecha = pd.DataFrame(columns=['FECHA', 'MUNICIPIO RESIDENCIA'], data=b)
final_delelegacion_fecha = final_delelegacion_fecha.sort_values(by=["FECHA"], ascending=True)
for delegacion in delegaciones:
    casos_delegacion = clean_covid_data.copy()
    casos_delegacion = casos_delegacion[casos_delegacion['MUNICIPIO RESIDENCIA']==delegacion]
    casos_delegacion = casos_delegacion.groupby(["FECHA INGRESO","MUNICIPIO RESIDENCIA"]).count().reset_index().rename(columns={'ID_REGISTRO':f'CASOS {delegacion}','FECHA INGRESO':'FECHA'}).drop(['SEXO','FECHA DEFUNCION','INTUBADO','EDAD','DIABETES','EPOC','ASMA','INMUNOSUPRESION','HIPERTENSION','CARDIOVASCULAR','OBESIDAD','RENAL CRONICA','TABAQUISMO'],axis=1)
    casos_delegacion = pd.merge(fecha, casos_delegacion,  how='left', left_on=['FECHA'], right_on = ['FECHA'])
    casos_delegacion['MUNICIPIO RESIDENCIA'] = delegacion
    casos_delegacion = casos_delegacion.fillna(0)
    casos_delegacion[f'ACUMULADO {delegacion}'] = casos_delegacion[F'CASOS {delegacion}'].cumsum(axis = 0)
    final_delelegacion_fecha = pd.merge(final_delelegacion_fecha, casos_delegacion,  how='left', left_on=['FECHA', 'MUNICIPIO RESIDENCIA'], right_on = ['FECHA', 'MUNICIPIO RESIDENCIA'])
    final_delelegacion_fecha = final_delelegacion_fecha.fillna(0)
list1 = []
final_delelegacion_fecha['ACUMULADO'] = 0
final_delelegacion_fecha['CASOS'] = 0
for delegacion in delegaciones:
    list1.append(f'CASOS {delegacion}')
    list1.append(f'ACUMULADO {delegacion}')
    final_delelegacion_fecha['ACUMULADO'] = final_delelegacion_fecha['ACUMULADO'] + final_delelegacion_fecha[f'ACUMULADO {delegacion}']
    final_delelegacion_fecha['CASOS'] = final_delelegacion_fecha['CASOS'] + final_delelegacion_fecha[f'CASOS {delegacion}']
final_delelegacion_fecha = final_delelegacion_fecha.drop(list1,axis=1)
final_delelegacion_fecha
```

```python
pd.DataFrame.to_csv(final_delelegacion_fecha,'../static/data/casos_delegacion_fecha.csv',encoding='latin1')
```
