```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
```

```python
file_to_load = "../../Resources/casos-asociados-a-covid-19.csv"

covid_cases_data = pd.read_csv(file_to_load)
covid_cases_data.head()
```

```python
clean_covid_data = covid_cases_data[covid_cases_data["Negativo"]!="No positivo SARS-CoV-2"]
clean_covid_data = clean_covid_data[["ID_REGISTRO","SEXO","MUNICIPIO RESIDENCIA","FECHA INGRESO", "FECHA DEFUNCION","INTUBADO","EDAD","DIABETES","EPOC","ASMA","INMUNOSUPRESION","HIPERTENSION","CARDIOVASCULAR","OBESIDAD","RENAL CRONICA","TABAQUISMO"]]
display(clean_covid_data)
```

```python
casos_delegacion = clean_covid_data.groupby(["MUNICIPIO RESIDENCIA"]).count().reset_index().rename(columns={'ID_REGISTRO':'CASOS','MUNICIPIO RESIDENCIA':'DELEGACION'}).drop(['SEXO','FECHA INGRESO','FECHA DEFUNCION','INTUBADO','EDAD','DIABETES','EPOC','ASMA','INMUNOSUPRESION','HIPERTENSION','CARDIOVASCULAR','OBESIDAD','RENAL CRONICA','TABAQUISMO'],axis=1)
display(casos_delegacion)
```

```python
pd.DataFrame.to_csv(casos_delegacion,'../assets/data/casos_por_delegacion.csv',encoding='latin1')
```
