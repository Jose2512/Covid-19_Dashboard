```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "../../Resources/casos-asociados-a-covid-19.csv"

# Read Purchasing File and store into Pandas data frame
covid_cases_data = pd.read_csv(file_to_load)
covid_cases_data.head()
```

```python
clean_covid_data = covid_cases_data[["ID_REGISTRO","SEXO","FECHA INGRESO", "FECHA DEFUNCION","INTUBADO","EDAD","DIABETES","EPOC","ASMA","INMUNOSUPRESION","HIPERTENSION","CARDIOVASCULAR","OBESIDAD","RENAL CRONICA","TABAQUISMO"]]
display(clean_covid_data)
```

```python
cases_per_date = clean_covid_data.copy()
cases_per_date = cases_per_date.groupby(['FECHA INGRESO']).count().rename(columns={'ID_REGISTRO':'CASOS'}).drop(['SEXO','FECHA DEFUNCION','INTUBADO','EDAD','DIABETES','EPOC','ASMA','INMUNOSUPRESION','HIPERTENSION','CARDIOVASCULAR','OBESIDAD','RENAL CRONICA','TABAQUISMO'],axis=1).reset_index()
cases_per_date
```

```python
pd.DataFrame.to_csv(cases_per_date,'../assets/data/casos_fecha_ingreso.csv',encoding='latin1')
```
